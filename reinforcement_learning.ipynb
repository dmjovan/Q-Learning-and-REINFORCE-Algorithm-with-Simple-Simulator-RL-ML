# Učitavanje biblioteka za rad 

```python
from qlearning import QLearningAgent
from reinforce import ReinforceAgent
```

# Okruženje

# Treniranje Q-Learning agenta

## Agent sa faktorom umanjenja buducih nagrada $\gamma$=0.9 i sa promenljivom stopom ucenja

```python
agent = QLearningAgent(gamma=0.9, print_infos=False)
agent.train()
agent.evaluate()
```

## Agent sa faktorom umanjenja buducih nagrada $\gamma$=0.9 i sa konstantnom stopom ucenja

```python
for alpha in [0.001, 0.005, 0.01, 0.05, 0.1, 0.2]:
    del agent
    agent = QLearningAgent(gamma=0.9, alpha=alpha, print_infos=False)
    agent.train()
    agent.evaluate()
```

## Agent sa faktorom umanjenja buducih nagrada $\gamma$=0.999 i sa promenljivom stopom ucenja

```python
del agent
agent = QLearningAgent(gamma=0.999, print_infos=False)
agent.train()
agent.evaluate()
```

## Agent sa faktorom umanjenja buducih nagrada $\gamma$=0.999 i sa konstantnom stopom ucenja

```python
for alpha in [0.001, 0.005, 0.01, 0.05, 0.1, 0.2]:
    del agent
    agent = QLearningAgent(gamma=0.999, alpha=alpha, print_infos=False)
    agent.train()
    agent.evaluate()
```

Primetno je da povecavanjem faktora umanjenja $\gamma$, imamo brzu i znacajno stabilniju konvergenciju algoritma, sto nije slucaj sa nesto manjim pomenutim faktorom. \\
Na osnovu prikazanih grafika, vidimo da je u slucaju veceg faktora umanjenja, najbolji metod da se stopa ucenja polako smanjuje ili da bude oko 0.2, jer u tim slucajevima imamo najfiniju i najbrzu konvergenciju. 
Za manji faktor umanjenja, najbolje je podesiti epizodno smanjivanje stop ucenja.\\

Eksperimentalno je odredjeno da verovatnoca $\epsilon$ treba da bude otprilike 0.5, 0.6 i da se smanjuje nakon svake epizode.

Takodje, primetno je da nijedna naucna politika nije ona najintuitivnija, a daje odlicne rezultate, sto znaci da je agent poprilicno dobro naucio da resi zadatak.

# Treniranje REINFORCE agenta

## Agent sa promenljivom stopom ucenja

```python
del agent
agent = ReinforceAgent()
agent.train_and_evaluate()
```

Agent sa promenljivom stopom ucenja daje u svim fazama evaluacije adekvatnu politiku, tj. politku koja je najintuitivnija, sto je odlicno. \\
Kod agent sa promenljivom stopom ucenja, imamo poprilicno stacionarne parametre za sva neterminalna stanja i akcije u toku ucenja, sto je takodje odlicno, jer nema znacajnih promena, vec je agent naucio dobro svoje parametre.\\
Prosecna ukupna nagrada je veoma zadovoljavajuca. 

## Agent sa konstantnom stopom ucenja

```python
for alpha in [0.001, 0.005, 0.01, 0.05, 0.1, 0.2, 0.5, 1]:
    print(f'-------------------------------------------------------------------------------------- alpha={alpha} --------------------------------------------------------------------------------------')
    del agent
    agent = ReinforceAgent(num_episodes=600, alpha=alpha)
    agent.train_and_evaluate()
```

Vec od stope ucenja 0.01, dobijamo u svakoj fazi evaluacije najbolju politiku, sto je odlican rezultat. \
Primetno je da najstacionarnije parametre agenta dobijamo povecavanjem stope ucenja, medjutim, prosecna ukupna nagrada opada sa broj epizoda, sto nije dobro. \
Kompromis za ovaj algoritam je ili resavanje problema sa promenljivom stopom ucenja, ili sa konstantnm stopom ucenja od otprilike 0.1-0.3.

Kako su gotovo sve politike ajintuitivije, a videli smo iz Q-Learning agenta da obrnuti slucaj daje odlicne rezultate, tako i ovde mozemo protumaciti manje prosecne ukupne nagrade kao posledica tako naucene politike. Naravno, ovako naucena politika dosta je vezana za odabir obelezja, koji je ovde bio skoncentrisan na jurenje nagrade i na izbegavanje "smrti".
